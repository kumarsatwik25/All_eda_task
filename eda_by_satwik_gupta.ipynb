### For credit score column

```python
#fiding IOR
per25=new_data["CreditScore"].quantile(0.25)
per75=new_data["CreditScore"].quantile(0.75)

```

```python
iqr=per75-per25
```

```python
upper_limit = per75 + 1.5 * iqr
lower_limit = per25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
new_data[new_data['CreditScore'] > lower_limit]
```

### Capping on the data

```python
new_data_cap = new_data.copy()

new_data_cap['CreditScore'] = np.where(
    new_data_cap['CreditScore'] > upper_limit,
    upper_limit,
    np.where(
        new_data_cap['CreditScore'] < lower_limit,
        lower_limit,
        new_data_cap['CreditScore']
    )
)
```

```python
new_data_cap
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(new_data['CreditScore'])

plt.subplot(2,2,2)
sns.boxplot(new_data['CreditScore'])

plt.subplot(2,2,3)
sns.distplot(new_data_cap['CreditScore'])

plt.subplot(2,2,4)
sns.boxplot(new_data_cap['CreditScore'])

plt.show()
```
